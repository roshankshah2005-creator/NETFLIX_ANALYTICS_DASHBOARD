Phase 2:**Data Cleaning**

```python
#Creating a copy of dataset
netflix=df.copy()
```

```python
print("Shape:",netflix.shape)
```

```python
netflix.info()
```

```python
missing=netflix.isnull().sum()
```

```python
missing
```

```python
missing_percentage = (netflix.isnull().sum()/len(netflix))*100

missing_percentage.sort_values(ascending=False)
```

```python
#Visualising missing data
missing_percentage = missing_percentage[missing_percentage > 0]
plt.figure(figsize=(10,5))
plt.bar(missing_percentage.index,missing_percentage.values)
plt.xticks(rotation=45)
plt.ylabel("Missing Percentage")

plt.title("Missing Values by Column")
plt.show()
```

```python
#Removing Duplicate data
print("Duplicate rows:",netflix.duplicated().sum())
netflix.drop_duplicates(inplace=True)
print("After dropping:",netflix.shape)
```

```python
netflix["date_added"] = pd.to_datetime(
    netflix["date_added"], 
    format="mixed"
)
```

```python
netflix["date_added"].head()
```

```python
netflix['year_added']=netflix['date_added'].dt.year
netflix['month_added']=netflix['date_added'].dt.month_name()
netflix['day_added']=netflix['date_added'].dt.day
```

```python
netflix[["date_added",
         "year_added",
         "month_added",
         "day_added"]].head()
```

```python
text_columns = [
    "director",
    "cast",
    "country",
    "rating"
]

for col in text_columns:
    netflix[col] = netflix[col].fillna("Unknown")
```

```python
netflix['description']=netflix['description'].fillna(" ")
```

```python
netflix.isnull().sum()
```

```python
netflix.dtypes
```

```python
netflix.to_csv("C:/Users/ROSHAN/Netflix-Analytics-Dashboard/data/netflix_titles_updated.csv",index=False)
```

Phase 3:**Exploratory Data Analysis**

##EDA 1: Movies vs TV Shows



##What percentage of Netflix content consists of Movies and TV Shows?

```python
typec=netflix['type'].value_counts()
print(typec)
```

```python
plt.figure(figsize=(12,6))
plt.pie(
    typec,
    labels=typec.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Movies vs TV shows on netflix')
plt.show()
```

Conclusion:Movies makes major of Netflix catalog,People prefer more movies rather than episodic content.

**EDA 2: Content Added Over the Years**

How has Netflix expanded its catalog over time?

```python
content_year=netflix['date_added'].value_counts().sort_index()
content_year
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    content_year.index,
    content_year.values,
    marker="o"
)

plt.title("Content Added to Netflix Over the Years")

plt.xlabel("Year")

plt.ylabel("Number of Titles")

plt.grid()

plt.show()
```

There was a huge growth during COVID during 2020 but soon after that the growth started declining.

**EDA 3: Top 10 Countries**

```python
country=netflix.copy()
```
